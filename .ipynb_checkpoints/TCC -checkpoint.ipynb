```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('20200617_tccfile.csv', sep =';')  
```

```python
pd.options.display.max_rows = 100
pd.options.display.max_columns  = 1000
```

```python
df
```

```python
cidades = df.columns[pd.Series(df.columns).str.startswith('Cidade')]
cidades
```

```python
def Columns (dataframe, columns, newcolumn):
    x=0
    dataframe[newcolumn] = ''
    while x <= len(columns)-1:
        dataframe[newcolumn]= (dataframe[newcolumn] + dataframe[columns[x]].fillna('') )
        x = x+1
    
    dataframe = dataframe.drop(columns, axis=1)
    return dataframe

```

```python
dataframe = Columns(df, cidades, 'cidades_')
```

```python
dataframe
```

```python
dataframe = dataframe.drop(['Blank [narrative_40]', 'Fim :)','Novo Cenario','Só falta um','Quase la o/'], axis=1)
```
